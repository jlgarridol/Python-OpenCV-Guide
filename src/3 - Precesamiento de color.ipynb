```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagenC = cv2.imread('images/imcl30.jpg',1)
imagenBN = cv2.imread('images/imcl30.jpg',0)

imagenC = cv2.cvtColor(imagenC,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.imshow(imagenC)
plt.title('Color')
plt.subplot(1,2,2)
plt.imshow(imagenBN,cmap='gray')
plt.title('Escala de grises')
```

```python
R,G,B = cv2.split(imagenC)
Z = np.zeros(R.shape,dtype='uint8')
R = cv2.merge((R,Z,Z))
G = cv2.merge((Z,G,Z))
B = cv2.merge((Z,Z,B))

H,S,V = cv2.split(cv2.cvtColor(imagenC,cv2.COLOR_RGB2HSV_FULL))
O = np.ones(H.shape,dtype='uint8')*255
H = cv2.cvtColor(cv2.merge((H,O,O)),cv2.COLOR_HSV2RGB_FULL)

plt.figure(figsize=(16,8))
plt.subplot(1,3,1)
plt.imshow(R)
plt.title('Rojo')
plt.subplot(1,3,2)
plt.imshow(G)
plt.title('Verde')
plt.subplot(1,3,3)
plt.imshow(B)
plt.title('Azul')

plt.figure(figsize=(16,8))
plt.subplot(1,3,1)
plt.imshow(H)
plt.title('Tono')
plt.subplot(1,3,2)
plt.imshow(S,cmap='gray')
plt.title('Saturación')
plt.subplot(1,3,3)
plt.imshow(V,cmap='gray')
plt.title('Valor')
```
